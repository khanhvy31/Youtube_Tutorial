```python
##Set up pandas SQL
query = '''
select a.*, sum(payment_value) as sum_total
 from order_data a
join payment b
on a.order_id = b.order_id
group by a.order_id
having sum_total >= 100
order by order_id
'''

df = ps.sqldf(query,locals() )
```
